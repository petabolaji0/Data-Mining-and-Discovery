```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Number of Employee
nemployee = 1000
```

```python
# Employee ID
np.random.seed(25)
employeeids = np.unique(np.random.randint(658000, 6581001, size=nemployee))
len(employeeids)
```

```python
# Names of Employee
names = np.loadtxt('/content/drive/MyDrive/DM & D/names.csv', delimiter=',',\
                            dtype=str) #https://www.name-generator.org.uk/quick/
len(names)
```

```python
# Postcode
postcodes = [f'EN{str(i).zfill(2)}' for i in range(1, 11)]
postcode= np.random.choice(postcodes, nemployee)
len(postcode)
```

```python
# Date of Employment
np.random.seed(25)
employment_year = np.random.randint(2001, 2020, nemployee)
employment_month = np.random.randint(1, 13, nemployee)
employment_day = np.random.randint(1, 31, nemployee)
employment_date = [f'{employment_year[i]}-{str(employment_month[i]).zfill(2)}-'\
                     f'{str(employment_day[i]).zfill(2)}' for i in range(nemployee)]
len(employment_date)
```

```python
# Gender
genders = ['M', 'F']
gender = np.random.choice(genders, size=nemployee)
len(gender)
```

```python
# Age
np.random.seed(25)
age = np.random.randint(21, 51, size=nemployee)
len(age)
```

```python
# Job Title
job_roles = {
    'Food Scientist': 20, 'Quality Assurance Specialist': 20,\
    'Production Manager': 40, 'Food Technologist': 20,\
    'Packaging Operator': 500, 'Supply Chain Coordinator': 50,\
    'Research and Development Chef': 50, 'Quality Control Inspector': 200,\
    'Logistics Coordinator': 50,'Maintenance Technician': 50\
    }

roles = list(job_roles.keys())
size = [job_roles[role] for role in roles]

job_title = np.random.choice(roles, nemployee, p=size/np.sum(size))
len(job_title)
```

```python
# Annaul Salary
np.random.seed(25)
salary_ranges = {
          'Food Scientist': 40000, 'Quality Assurance Specialist': 40000,\
          'Production Manager': 50000, 'Food Technologist': 32000,\
          'Packaging Operator': 30000, 'Supply Chain Coordinator': 35000,\
          'Research and Development Chef': 42000,'Quality Control Inspector': 35000,\
          'Logistics Coordinator': 40000, 'Maintenance Technician': 40000\
          }

annual_salary = np.array([np.random.randint(salary_ranges[job],\
                                            salary_ranges[job]+1000) for job in job_title])
len(annual_salary)
```

```python
# Performance Rating
np.random.seed(5)
ratings = ['Excellent', 'Good', 'Fair']
performance_rating = np.random.choice(ratings, size=nemployee)

len(performance_rating)
```

```python
# Create DataFrame
employee_db = pd.DataFrame({
    'Employee_ID': employeeids,\
    'Names': names,\
    'Postcode': postcode,\
    'Gender': gender,\
    'Age': age,\
    'Employment_Date':employment_date,\
    'Job_Title': job_title,\
    'Annual_Salary': annual_salary,\
    'Performance_Rating': performance_rating
    })

employee_db.set_index('Employee_ID', inplace=True)
employee_db
```

```python
employee_db.info()
```

```python
# Convert to .csv file
employee_db.to_csv('EmployeeData.csv')
employee_db[['Names', 'Postcode', 'Gender', 'Age']].to_csv('EmployeeDetails.csv')
employee_db[['Employment_Date']].to_csv('EmploymentDate.csv')
employee_db[['Job_Title']].to_csv('JobTitle.csv')
employee_db[['Annual_Salary']].to_csv('AnnualSalary.csv')
employee_db[['Performance_Rating']].to_csv('Performance.csv')
```
